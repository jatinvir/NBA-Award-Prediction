```python
import pandas as pd
```

```python
mvps = pd.read_csv('mvps.csv')
```

```python
mvps.head()
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
mvps
```

```python
players = pd.read_csv('players.csv')
```

```python
players
```

```python
del players['Unnamed: 0']
del players['Rk']
```

```python
players.head()
```

```python
players["Player"] = players["Player"].str.replace("*", "")
```

```python
players.head(50)
```

```python
def single_player(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
players.reset_index(inplace=True)

players = players.groupby(["Player", "Year"]).apply(single_player)
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
teams = pd.read_csv('teams.csv')
```

```python
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams["Team"].unique()
```

```python
combined["Tm"].unique()
```

```python
nicknames = {}

with open("nicknames.txt") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name =  line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
stats = combined.merge(teams, how="outer", on=["Team", "Year"])
```

```python
del stats["Unnamed: 0"]
```

```python
stats = stats.apply(pd.to_numeric, errors='ignore')
```

```python
stats["GB"] = stats["GB"].str.replace("—", "0")
stats['GB'].unique()
```

```python
stats["GB"] = pd.to_numeric(stats["GB"])
```

```python
stats.to_csv("player_mvp_stats.csv", index=False)
```

```python
highest_scoring = stats[stats["G"] > 70].sort_values("PTS", ascending=False).head(10)
```
